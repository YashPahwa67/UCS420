```python
import pandas as pd

employee_data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "FirstName": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [75000, 85000, 70000, 65000, 60000],
    "YearsOfExperience": [4, 8, 10, 3, 12],
    "JoiningDate": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "male", "male", "Female", "male"],
    "Bonus": [4.5, 4.0, 3.8, 4.7, 3.5]
}
df = pd.DataFrame(employee_data)

df.to_csv("employees.csv", index=False)

print("employees.csv file has been created successfully!")
df.head()
```

```python
shape = df.shape
print(f"The shape of the DataFrame is: {shape}")
print(f"Number of rows: {shape[0]}")
print(f"Number of columns: {shape[1]}")
```
