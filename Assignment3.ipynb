6. Create a sample dataset (employees.csv) containing information about employees in
a company.
---------------------------------------------------------------------------------------------------------------------------------------------

```python
import pandas as pd

data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': pd.to_datetime(['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25']),
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)

print("Shape of DataFrame:", df.shape)

print("\nSummary of DataFrame:")
print(df.info())

print("\nDescriptive Statistics:")
print(df.describe())

print("\nFirst 5 rows:")
print(df.head())
print("\nLast 3 rows:")
print(df.tail(3))


print("\nAverage Salary:", df['Salary'].mean())
print("Total Bonus Paid:", df['Bonus'].sum())
print("Youngest Employee Age:", df['Age'].min())
print("Highest Performance Rating:", df['Rating'].max())

df_sorted = df.sort_values(by='Salary', ascending=False)
print("\nDataFrame sorted by Salary:")
print(df_sorted)

def categorize_performance(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Performance_Category'] = df['Rating'].apply(categorize_performance)
print("\nDataFrame with Performance Category:")
print(df)

print("\nMissing Values:")
print(df.isnull().sum())

df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\nDataFrame with renamed column:")
print(df)

experienced_employees = df[df['Years_of_Experience'] > 5]
print("\nEmployees with more than 5 years of experience:")
print(experienced_employees)

it_employees = df[df['Department'] == 'IT']
print("\nEmployees in IT department:")
print(it_employees)

df['Tax'] = df['Salary'] * 0.10
print("\nDataFrame with Tax column:")
print(df)

df.to_csv("modified_employees.csv", index=False)
print("\nModified DataFrame saved to 'modified_employees.csv'")
```
